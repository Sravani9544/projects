```python

counts, bin_edges = np.histogram(haberman['axil_nodes'], bins=10, 
                                 density = True)
pdf = counts/(sum(counts))
print(pdf);
print(bin_edges);
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf);
plt.plot(bin_edges[1:], cdf)


counts, bin_edges = np.histogram(haberman['axil_nodes'], bins=20, 
                                 density = True)
pdf = counts/(sum(counts))
plt.plot(bin_edges[1:],pdf);

plt.show();


```

```python

counts, bin_edges = np.histogram(haberman['operation_year'] , bins = 20, density = True)
pdf = counts/(sum(counts))
print(pdf);
print(bin_edges)
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)
plt.show();


# the patient had equal chance of survival as the heights of pdf are almost equal 
# as seen in case of histograms the operation year does not exactly differentiate between survival or death of a patient
```

```python

counts, bin_edges = np.histogram(haberman['age'], bins = 20, density = True)
pdf = counts/(sum(counts))
print(pdf);
print(bin_edges)
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)
plt.show();

# the patients below 58 years of age have fairly good chance of survival of around 60 %
# the patients above 58 years have less chances of survival of more than 5 years 
```

```python

counts, bin_edges = np.histogram(haberman['axil_nodes'] , bins = 10, density = True)
pdf = counts/(sum(counts))
print(pdf);
print(bin_edges)
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)
plt.show();

# the patients having less than 30 positive axillary nodes have 60% of survival chances more than 5 years
# the patients having more than 30 positive axillary nodes have only 40% of survival chances more yhan 5 years
```

```python
import seaborn as sns
sns.set_style("whitegrid");
sns.FacetGrid(haberman,hue="surv_status" ,size = 6)\
   .map(plt.scatter, "axil_nodes", "operation_year")\
   .add_legend();
plt.show();

# this plot is a 2d scatter plot of axiallary nodes vs operation year 
# though it is not linearly seperable , there is not much concentration of any one point
# thus survival status does not depend upon operation year but the patients may survive if they have less than 10 axillary nodes
```
